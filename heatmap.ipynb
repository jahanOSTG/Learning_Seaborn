```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("/content/drive/MyDrive/boston_house_prices.xlsx - boston_house_prices.csv")
print(data)
```

```python
print(data.head())
```

```python
corr = data.corr(numeric_only=True)

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap - House Dataset")
plt.show()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5, linecolor='gray', mask=mask)
plt.title("Correlation Heatmap (Upper Triangle Masked)")
plt.show()

```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, center=0, cmap="vlag", linewidths=0.5)
plt.title("Centered Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis", yticklabels=False)
plt.title("Missing Values in Dataset")
plt.show()
```
