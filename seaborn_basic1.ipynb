```python
import seaborn as sns
import matplotlib.pyplot as plt
data=sns.load_dataset("tips")
print(data.head())
```

```python
plt.figure(figsize=(5,5))
sns.barplot(data=data,x='size',y='total_bill')
plt.title("Total_bill vs size")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.boxplot(data=data,x='size',y='total_bill',color='hotpink')
plt.title("Total_bill vs size")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=data, x="day", y="total_bill",  split=True)
plt.title("Total Bill by Day and Gender")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(data=data, x="day", y="total_bill",  split=False)
plt.title("Total Bill by Day and Gender")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.violinplot(data=data, x="day", y="total_bill", hue="sex", split=True)
plt.title("Total Bill by Day and Gender")
plt.show()
```
